```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt


PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "data").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

FEATURE_DIR = PROJECT_ROOT / "data" / "processed" / "features"


df_sectors = pd.read_csv(FEATURE_DIR / "sector_returns_processed.csv", index_col=0, parse_dates=True)
df_spx = pd.read_csv(FEATURE_DIR / "spx_processed.csv", index_col=0, parse_dates=True)


print("First 5 rows of Sector Returns:")
display(df_sectors.head())


print("\nSummary Statistics for Sectors:")
display(df_sectors.describe())
```

```python

plt.figure(figsize=(10, 6))

df_sectors['return'].dropna().hist(bins=100, color='skyblue', edgecolor='black')

plt.title('Distribution of Daily Sector Returns (2000 - Present)')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```
