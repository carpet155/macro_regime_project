All of the data shares the same 'observation_date' column which can be helpful for merging. Some are monthly values, some are daily values, most of the daily values won't have Saturdays or Sundays so there will be 2 dates gone for every 7, I'm guessing that holidays are also left off so we could also just remove the nulls since they're mostly holidays.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
CPI = pd.read_csv('../data/raw/CPIAUCSL.csv')
DGS2 = pd.read_csv('../data/raw/DGS2.csv')
DGS10 = pd.read_csv('../data/raw/DGS10.csv')
FEDFUNDS = pd.read_csv('../data/raw/FEDFUNDS.csv')
SP500 = pd.read_csv('../data/raw/SP500.csv')
VIXCLS = pd.read_csv('../data/raw/VIXCLS.csv')
```

```python
CPI['CPIAUCSL'] = pd.to_numeric(CPI['CPIAUCSL'], errors='coerce')
CPI['CPIAUCSL'] = CPI['CPIAUCSL'].interpolate()
```

```python
DGS2['DGS2'] = pd.to_numeric(DGS2['DGS2'], errors='coerce')
DGS2['DGS2'] = DGS2['DGS2'].interpolate()
DGS2.isna().sum()
```

```python
DGS10['DGS10'] = pd.to_numeric(DGS10['DGS10'], errors='coerce')
DGS10['DGS10'] = DGS10['DGS10'].interpolate()
DGS10.isna().sum()
```

```python
SP500 = SP500.iloc[2:]

# rename columns
SP500 = SP500[['Date', 'Adj Close']].rename(columns={
    'Date': 'observation_date',
    'Adj Close': 'SP500'
})

# make SP500 numeric and round to 2 decimals
SP500["SP500"] = pd.to_numeric(SP500["SP500"], errors="coerce").round(2)

# make observation_date a string
SP500["observation_date"] = SP500["observation_date"].astype(str)

SP500.head()
```

```python
VIXCLS['VIXCLS'] = pd.to_numeric(VIXCLS['VIXCLS'], errors='coerce')
VIXCLS['VIXCLS'] = VIXCLS['VIXCLS'].interpolate()
VIXCLS.isna().sum()
```
