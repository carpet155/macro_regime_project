```python
import pandas as pd
import os
print(os.getcwd())
```

```python
processed_dir = "../data/processed"

inflation = pd.read_csv(f"{processed_dir}/inflation_processed.csv", parse_dates=["date"])
inflation = inflation.rename(columns={"value": "cpi"})

treasury = pd.read_csv(f"{processed_dir}/treasury_processed.csv", parse_dates=["date"])

vix = pd.read_csv(f"{processed_dir}/vix_processed.csv", parse_dates=["date"])

spx = pd.read_csv(f"{processed_dir}/spx_processed.csv", parse_dates=["date"])
spx = spx.rename(columns={"price": "spx_price", "return": "spx_return"})

sectors = pd.read_csv(f"{processed_dir}/sectors_processed.csv", parse_dates=["date"])
sectors = sectors.rename(columns={"price": "sector_price", "return": "sector_return"})
```

```python
macro = inflation.merge(treasury, on="date", how="outer")
macro = macro.merge(vix, on="date", how="outer")
macro = macro.merge(spx, on="date", how="outer")
macro = macro.sort_values("date").reset_index(drop=True)

macro.head()
```

```python
master = sectors.merge(macro, on="date", how="left")
master = master.sort_values(["date", "ticker"]).reset_index(drop=True)

master.head()
```

```python
print(f"Shape: {master.shape}")
print(f"Date range: {master['date'].min().date()} → {master['date'].max().date()}")
print(f"Unique tickers: {master['ticker'].nunique()}")
print(f"Duplicates: {master.duplicated(subset=['date', 'ticker']).sum()}")
print("\nMissing values:")
print(master.isnull().sum())
```

```python
master.to_csv(f"{processed_dir}/master_df.csv", index=False)
print("Saved to data/processed/master_df.csv")
```
