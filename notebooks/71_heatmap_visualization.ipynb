# Heatmap Visualization

This notebook creates heatmaps showing sector returns across different macro regimes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)
```

```python
# Load the data
df_combined = pd.read_csv('../data/processed/pivot_sector_by_combined_regime.csv', index_col=0)
df_vix = pd.read_csv('../data/processed/pivot_sector_by_vix_regime.csv', index_col=0)
df_inflation = pd.read_csv('../data/processed/pivot_sector_by_inflation_regime.csv', index_col=0)
df_rate = pd.read_csv('../data/processed/pivot_sector_by_rate_regime.csv', index_col=0)

print("Data loaded successfully!")
print(f"\nCombined Regime data shape: {df_combined.shape}")
print(f"VIX Regime data shape: {df_vix.shape}")
print(f"Inflation Regime data shape: {df_inflation.shape}")
print(f"Rate Regime data shape: {df_rate.shape}")
```

## Heatmap 1: Sector Returns by Combined Macro Regime

Shows average daily returns for each sector under different macro regimes (based on both inflation and interest rate environments).

```python
plt.figure(figsize=(14, 5))
sns.heatmap(df_combined, 
            annot=True, 
            fmt='.4f', 
            cmap='RdYlGn', 
            center=0,
            cbar_kws={'label': 'Average Daily Return'},
            linewidths=0.5,
            linecolor='gray')
plt.title('Sector Returns by Combined Macro Regime', fontsize=14, fontweight='bold')
plt.xlabel('Sector', fontsize=12)
plt.ylabel('Regime', fontsize=12)
plt.tight_layout()
plt.show()
```

## Heatmap 2: Sector Returns by VIX Regime

Shows how sector performance differs during periods of market calm vs. stress (based on VIX levels).

```python
plt.figure(figsize=(14, 3))
sns.heatmap(df_vix, 
            annot=True, 
            fmt='.4f', 
            cmap='RdYlGn', 
            center=0,
            cbar_kws={'label': 'Average Daily Return'},
            linewidths=0.5,
            linecolor='gray')
plt.title('Sector Returns by VIX Regime (Market Volatility)', fontsize=14, fontweight='bold')
plt.xlabel('Sector', fontsize=12)
plt.ylabel('Regime', fontsize=12)
plt.tight_layout()
plt.show()
```

## Heatmap 3: Sector Returns by Inflation Regime

```python
plt.figure(figsize=(14, 4))
sns.heatmap(df_inflation, 
            annot=True, 
            fmt='.4f', 
            cmap='RdYlGn', 
            center=0,
            cbar_kws={'label': 'Average Daily Return'},
            linewidths=0.5,
            linecolor='gray')
plt.title('Sector Returns by Inflation Regime', fontsize=14, fontweight='bold')
plt.xlabel('Sector', fontsize=12)
plt.ylabel('Regime', fontsize=12)
plt.tight_layout()
plt.show()
```

## Heatmap 4: Sector Returns by Interest Rate Regime

```python
plt.figure(figsize=(14, 4))
sns.heatmap(df_rate, 
            annot=True, 
            fmt='.4f', 
            cmap='RdYlGn', 
            center=0,
            cbar_kws={'label': 'Average Daily Return'},
            linewidths=0.5,
            linecolor='gray')
plt.title('Sector Returns by Interest Rate Regime', fontsize=14, fontweight='bold')
plt.xlabel('Sector', fontsize=12)
plt.ylabel('Regime', fontsize=12)
plt.tight_layout()
plt.show()
```

## Summary

The heatmaps above show average daily sector returns across different macro regimes:

- **Green cells**: Positive returns
- **Red cells**: Negative returns
- **Lighter colors**: Values closer to zero

Key insights from the visualizations:
- Different sectors perform differently depending on the macro regime
- Some sectors are more resilient during market stress (low VIX returns during stress regime)
- Interest rate and inflation regimes have distinct patterns of sector performance
